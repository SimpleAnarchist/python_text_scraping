# 2. Vikipedi Üzerinden Metinleri Türkçe Kazıma

```python
import requests
from bs4 import BeautifulSoup
```

```python
baslik = input("Başlığı Giriniz: ")
link = "https://tr.wikipedia.org/wiki/" + str(baslik.lower())
r = requests.get(link)
soup = BeautifulSoup(r.content, "lxml")
st1 = soup.find_all("div", attrs = {"id":"bodyContent"})
text = st1[0].text
print(text)
```

# 3. Haber Sitesinden Metin Kazıma

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
source = urlopen("https://www.hurriyet.com.tr/teknoloji/dijital-tuketiciler-2021-ekonomisine-yon-verecek-41685236").read()
soup = BeautifulSoup(source, "lxml")
print(set([text.parent.name for text in soup.find_all(text=True)]))
```

```python
text = ""

for paragraph in soup.find_all("h2"):
    text += paragraph.text
    
print(text)
```

```python
text = ""

for paragraph in soup.find_all("p"):
    text += paragraph.text
    
print(text)
```
